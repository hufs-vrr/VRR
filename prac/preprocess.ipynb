```python
import librosa
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt
import sys
import tensorflow as tf

samples, sample_rate = librosa.load(path= 'believer.wav', sr = 44100, mono = True, duration = 18.5)

fs_ = 44100
nperseg_ = fs_*0.23
noverlap_ = nperseg_*0.5


freq_axis, t_axis, spec = signal.spectrogram(x = samples, fs = 44100, nperseg = 1014, noverlap = 507, nfft= 1024)
#nfft가 nperseg보다 커야함

print(sample_rate)
print(len(spec),len(spec[0]),len(spec[1])) #2열(세로, 각각행의 값들)크기는 duration에 따라 달라짐, 1행크기는 nfft의 절반 정도
                                           #그리고 nfft는 windowsize에 영향을 받는다.
print(spec[0])
np.set_printoptions(threshold = sys.maxsize)
#f = open('4. input.txt', 'w')
new_spec = spec[0:512]
new_spec = np.transpose(new_spec)
new_spec = new_spec[0:1600]
#f.write(str(new_spec))
#f.close()
print(len(new_spec),len(new_spec[0]))
#amp = 2 * np.sqrt(2)
plt.pcolormesh(t_axis, freq_axis, np.abs(spec))
plt.title('STFT Magnitude')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```
